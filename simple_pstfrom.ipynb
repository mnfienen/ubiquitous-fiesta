```python
import os
import pyemu
import flopy
from useful_fxns import *
```

```python
org_ws = 'monthly_model_files_1lyr_newstress'
new_ws = 'simple_pst'
```

```python
if os.path.exists(new_ws):
    shutil.rmtree(new_ws)
shutil.copytree(org_ws, new_ws)
```

```python
sim = flopy.mf6.MFSimulation.load(sim_ws = new_ws)
```

```python
prep_deps(new_ws)
```

```python
pyemu.os_utils.run('mf6',cwd=new_ws)
```

```python
pf = pyemu.utils.PstFrom(new_ws,'template',remove_existing=True)
```

```python
pf.add_parameters('freyberg6.npf_k_layer1.txt',par_type='grid',upper_bound=20,lower_bound=0.05)
```

```python
pf.add_observations('heads.csv',index_cols='time',use_cols='TRGW-0-2-15')
```

```python
pf.mod_sys_cmds.append('mf6')
```

```python
pf.build_pst(filename=os.path.join(pf.new_d,'freyberg.pst'),version=2)
```

```python
pst = pyemu.Pst(filename=os.path.join(pf.new_d,'freyberg.pst'))
obs = pst.observation_data
obs.iloc[0,1] += 10
obs
pst.write(os.path.join(pf.new_d,'freyberg.pst'),version=2)
```
